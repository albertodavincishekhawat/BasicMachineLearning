```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import make_blobs, make_circles
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split

import warnings 
warnings.filterwarnings("ignore")
```

```python
x1, y1 = make_blobs(n_samples=50, centers=2, cluster_std=4, random_state=43)
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(x1[y1==0, 0], x1[y1==0, 1])
plt.scatter(x1[y1==1, 0], x1[y1==1, 1])
plt.show()
```

```python
X_tr, X_te, y_tr, y_te = train_test_split(x1, y1, train_size=0.8, random_state=33)
```

```python
model = SVC(kernel="linear")
```

```python
model.fit(X_tr, y_tr)
```

```python
model.score(X_tr, y_tr)
```

```python
y_pr = model.predict(X_te)
```

```python
accuracy_score(y_te, y_pr)
```

```python
x2, y2 = make_circles(n_samples=100, noise=0.1, factor=0.4)
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(x2[y2==0, 0], x2[y2==0, 1])
plt.scatter(x2[y2==1, 0], x2[y2==1, 1])
plt.show()
```

```python
X_tr, X_te, y_tr, y_te = train_test_split(x2, y2, train_size=0.8, random_state=33)
```

```python
model2 = SVC(kernel="linear")
```

```python
model2.fit(X_tr, y_tr)
```

```python
model2.score(X_tr, y_tr)
```

```python
y_pr = model2.predict(X_te)
```

```python
accuracy_score(y_te, y_pr)
```

```python
model3.fit(X_tr, y_tr)
```

```python
model3.score(X_tr, y_tr)
```

```python
accuracy_score(y_te, model3.predict(X_te))
```
