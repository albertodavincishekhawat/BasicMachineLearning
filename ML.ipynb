## Mainly there are 3 Types of Machine Learning :


#### Supervised
- Data / Input / Feature and Label / Output / Target
- Data can have n-number of columns but Label will have only one column
> Old car price prediction
- Feature : age,maker , model, seat, abs, features
- Target : Price

> Loan approver
- Feature : Education, Age, Credit, Account, Reside, Area
- Target : Approve/Disapprove

> Basically we teach the machine by providing example i.e, Feature and Target



#### Unsupervised
- Only Data / Input / Feature is provided and we find pattern by simillarity measures.
- We give input and machine tells us what the pattern is.

> we are giving data of customers of D-Mart to machine and it returns categories of customers

> A child segregating chocolates, chips and biscuits on it's own.




#### Reinforcment

- Data and Label both are given to the machine and it makes changes according to the feedback, judging on it's own output or feed back from user
